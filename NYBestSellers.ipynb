```python
import requests
import json
import time  #library used to get current time
import pandas as pd
```

```python
#create a text file that contains your API key
#you DO NOT want others to see your API key
#this method is not the most secure but will work for this exercise

#this will read the file and store your API key as a variable
with open('books_key.txt', 'r') as file:
    for line in file:
        api_key = line
```

```python
#sandbox database to practice getting data from their API
url = r"https://api.nytimes.com/svc/books/v3/lists/current/hardcover-fiction.json?api-key="
```

```python
#in the request to the website, give it the url and add the API key at the end
response = requests.get(url+api_key)
```

```python
#status code is 200, so this is good
response
```

```python
#this is the JSON string data from the API
response.text
```

```python
#set the string data to a variable
str_books = response.text
```

```python
#load in the string data as JSON
bookdata = json.loads(str_books)
```

```python
#create empty lists to extract data

primaryISBN10ls = [] 
primaryISBN13ls = [] 
titlels = [] 
authorls = []
rankls = [] 
lastweeksrankls = [] 
numweekslistls = [] 
pricels = [] 
descriptls = []
publisherls = [] 
imagels = [] 
```

```python
#set a "timer" for 3 sec after the code starts
end = time.time() + 3 

#while the current time is less than the ending time (3 sec after the code started)
while time.time() <= end:  
    for book in books:
        
        #get the value of each key and if the key doesn't exist, set a variable to be None
        try: primaryISBN10 = book['primary_isbn10']
        except: primaryISBN10 = None
            
        try: primaryISBN13 = book['primary_isbn13']
        except: primaryISBN13 = None
            
        try: title = book['title']
        except: title = None
            
        try: author = book['author']
        except: author = None
            
        try: rank = book['rank']
        except: rank = None
            
        try: lastweeksrank = book['rank_last_week']
        except: lastweeksrank = None
            
        try: numweekslist = book['weeks_on_list']
        except: numweekslist = None
            
        try: price = book['price']
        except: price = None
            
        try: descript = book['description']
        except: descript = None
            
        try: publisher = book['publisher']
        except: publisher = None
            
        try: image = book['book_image']
        except: image = None
            
            
        
        primaryISBN10ls.append(primaryISBN10)
        primaryISBN13ls.append(primaryISBN13)
        titlels.append(title)
        authorls.append(author)
        rankls.append(rank)
        lastweeksrankls.append(lastweeksrank)
        numweekslistls.append(numweekslist)
        pricels.append(price)
        descriptls.append(descript)
        publisherls.append(publisher)
        imagels.append(image)
```

```python
#match indices of each list and zip into one list
bookList = list(zip(primaryISBN10ls, primaryISBN13ls, titlels, 
                    authorls, rankls, lastweeksrankls, numweekslistls, 
                    pricels, descriptls, publisherls, imagels))

#names for columns in data frame
colnames = ['Primary_ISBN10', 'Primary_ISBN13', 'title', 'author', 'rank', 'last_weeks_rank', 
            'num_weeks_on_list', 'price', 'description', 'publisher', 
            'image']
```

```python
#create data frame with column names in it
df = pd.DataFrame(bookList, columns=colnames)

df.head(10)
```
